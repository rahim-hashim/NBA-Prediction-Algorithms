```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

def gamelog_scraper(gamelogs)
  for url_year in gamelogs:
    url = "https://www.basketball-reference.com/players/b/bryanko01/gamelog/1997"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find('table', {'id': 'pgl_basic'})
    # print the table headers
    table_header = table.find('thead')
    header_cols = table_header.find_all('th')
    header_names = []
    for h_index, header in enumerate(header_cols):
      if u'\xa0' in header.get_text():
        if h_index < 7: # '' -> LOC
          header_names.append('LOC')
        else: # '' -> WIN
          header_names.append('WIN')
      else:
        header_names.append(header.get_text())
    df = pd.DataFrame(columns=header_names[1:])

    table_rows = table.find_all('tr')
    for tr in table_rows:
      td = tr.find_all('td')
      row = [i.text for i in td]
      if len(row) > 10: # fix for 'Did not Play'
        df_row = pd.DataFrame([row], columns=header_names[1:])
        df = pd.concat([df, df_row], ignore_index=True)
  return df 
df.head(50)
```
